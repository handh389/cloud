```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
result = pd.read_csv('/content/drive/MyDrive/cbe_paper/result.csv', header = None)
result
```

```python
iteration = []

w_da_map = []
w_da_precision = []
w_da_recall = []
w_da_f1 = []
w_da_iou = []
w_da_apmc = []

wo_da_map = []
wo_da_precision = []
wo_da_recall = []
wo_da_f1 = []
wo_da_iou = []
wo_da_apmc = []

for i in range(10):
  iteration.append(float(result[i+1][0]))
  w_da_map.append(float(result[i+1][1][:-1]))
  w_da_precision.append(float(result[i+1][2][:-1]))
  w_da_recall.append(float(result[i+1][3][:-1]))
  w_da_f1.append(float(result[i+1][4][:-1]))
  w_da_iou.append(float(result[i+1][5][:-1]))
  w_da_apmc.append(float(result[i+1][6][:-1]))
  wo_da_map.append(float(result[i+1][7][:-1]))
  wo_da_precision.append(float(result[i+1][8][:-1]))
  wo_da_recall.append(float(result[i+1][9][:-1]))
  wo_da_f1.append(float(result[i+1][10][:-1]))
  wo_da_iou.append(float(result[i+1][11][:-1]))
  wo_da_apmc.append(float(result[i+1][12][:-1]))
```

```python
fig = plt.figure(figsize=(15,15))
plt.plot(iteration, w_da_map, 'o-', color = 'r', linewidth = 5, markersize = 10)
plt.plot(iteration, w_da_precision, 'o-', color = 'b', linewidth = 5, markersize = 10)
plt.plot(iteration, w_da_recall, 'o-', color = 'm', linewidth = 5, markersize = 10)
plt.plot(iteration, w_da_f1, 'o-', color = 'g', linewidth = 5, markersize = 10)
plt.plot(iteration, w_da_iou, 'o-', color = 'c', linewidth = 5, markersize = 10)
plt.plot(iteration, w_da_apmc, 'o-', color = 'y', linewidth = 5, markersize = 10)
plt.plot(iteration, wo_da_map, 'o:', color = 'r')
plt.plot(iteration, wo_da_precision, 'o:', color = 'b')
plt.plot(iteration, wo_da_recall, 'o:', color = 'm')
plt.plot(iteration, wo_da_f1, 'o:', color = 'g')
plt.plot(iteration, wo_da_iou, 'o:', color = 'c')
plt.plot(iteration, wo_da_apmc, 'o:', color = 'y')
plt.xlim([1000, 10000])
plt.legend(['mAP', 'Precision', 'Recall', 'F1-score', 'Avg IoU', 'AP of Marangini Cell'], fontsize = 25)
plt.text(6100, 68, 'Solid line: w/ Data Augmentation', fontsize = 20)
plt.text(6100, 66, 'Dotted line: w/o Data Augmentation', fontsize = 20)
plt.xlabel('Iteration Number', fontsize = 30, fontweight = 'bold')
plt.ylabel('%', fontsize = 30, fontweight = 'bold')
plt.xticks(fontsize = 30)
plt.yticks(fontsize = 30)
plt.show()
```
