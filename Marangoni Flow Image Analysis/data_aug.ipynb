```python
import imgaug.augmenters as iaa;
import cv2;
import os;
import random;
import glob;

#---------------------------------------------------------------------------------------------------------
# 파일 로딩
path = '/content/drive/MyDrive/cbe_paper/rod_particle_marangoni_bbx/*.jpg'

# 각 데이터 증강 기법당 만들 사진의 갯수
n = 20
#--------------------------------------------------------------------------------------------------------


#%% 색깔 바꾸기(Gamma contrast, 감마보정)
for img in range(n):
  title = random.choice(glob.glob(path));
  image = cv2.imread(title);
  lines = open(str(title[:-4]+ '.txt'), 'r');
  aug = iaa.GammaContrast((0.5, 2.0), per_channel=True);
  image_aug = aug(image = image);
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'gamma_contrast_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'gamma_contrast_'+str(img) +'.txt', 'w');
  txt.write(lines.read());
  lines.close();
  txt.close();
print('감마 보정 완료')

#%% 색깔 바꾸기(log contrast)
for img in range(n):
  title = random.choice(glob.glob(path));
  image = cv2.imread(title);
  lines = open(str(title[:-4]+ '.txt'), 'r');
  aug = iaa.LogContrast(gain=(0.6, 1.4), per_channel=True);
  image_aug = aug(image = image);
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'log_contrast_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'log_contrast_'+str(img) +'.txt', 'w');
  txt.write(lines.read());
  lines.close();
  txt.close();
print('로그 보정 완료')

#%% 반대 픽셀(inverse pixel)
titles = random.sample(glob.glob(path), n);
for img in range(n):
  title = titles[img];
  image = cv2.imread(title);
  lines = open(str(title[:-4]+ '.txt'), 'r');
  aug = iaa.Invert(1.0);
  image_aug = aug(image = image);    
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'invert_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'invert_'+str(img) +'.txt', 'w');
  txt.write(lines.read());
  lines.close();
  txt.close();
print('반대 픽셀 완료')


#%% 블러(Gaussian blur)
titles = random.sample(glob.glob(path), n);
for img in range(n):
  title = titles[img];
  image = cv2.imread(title);
  lines = open(str(title[:-4]+ '.txt'), 'r');
  aug = iaa.GaussianBlur(sigma=(0.8, 3.0));
  image_aug = aug(image = image);    
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'gaussian_blur_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'gaussian_blur_'+str(img) +'.txt', 'w');
  txt.write(lines.read());
  lines.close();
  txt.close();
print('가우시언 블러 완료')


#%% 블러(avg blur)
titles = random.sample(glob.glob(path), n);
for img in range(n):
  title = titles[img];
  image = cv2.imread(title);
  lines = open(str(title[:-4]+ '.txt'), 'r');
  aug = iaa.AverageBlur(k=(2, 11));
  image_aug = aug(image = image);    
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'avg_blur_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'avg_blur_'+str(img) +'.txt', 'w');
  txt.write(lines.read());
  lines.close();
  txt.close();
print('평균 블러 완료')


#%% 노이즈 만들기(pepper)
for img in range(n):
  title = random.choice(glob.glob(path));
  image = cv2.imread(title);
  lines = open(str(title[:-4]+ '.txt'), 'r');
  aug = aug = iaa.Pepper(0.1);
  image_aug = aug(image = image);
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'pepper_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'pepper_'+str(img) +'.txt', 'w');
  txt.write(lines.read());
  lines.close();
  txt.close();
print('후추 노이즈 완료')


#%% 노이즈 만들기(salt)
for img in range(n):
  title = random.choice(glob.glob(path));
  image = cv2.imread(title);
  lines = open(str(title[:-4]+ '.txt'), 'r');
  aug = iaa.Salt(0.1);
  image_aug = aug(image = image);
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'salt_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'salt_'+str(img) +'.txt', 'w');
  txt.write(lines.read());
  lines.close();
  txt.close();
print('소금 노이즈 완료')

#%% 그림 가리기(coarse dropout-10%)
for img in range(n):
  title = random.choice(glob.glob(path));
  image = cv2.imread(title);
  lines = open(str(title[:-4]+ '.txt'), 'r');
  aug = iaa.CoarseDropout(0.1, size_percent=0.1, per_channel=0.5);
  image_aug = aug(image = image);
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'coarse_dropout_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'coarse_dropout_'+str(img) +'.txt', 'w');
  txt.write(lines.read());
  lines.close();
  txt.close();
print('일반 가리기 완료')


#%% 그림 가리기(coarse pepper-10%)
for img in range(n):
  title = random.choice(glob.glob(path));
  image = cv2.imread(title);
  lines = open(str(title[:-4]+ '.txt'), 'r');
  aug = iaa.CoarsePepper(0.1, size_percent=(0.01, 0.1));
  image_aug = aug(image = image);
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'coarse_pepper_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'coarse_pepper_'+str(img) +'.txt', 'w');
  txt.write(lines.read());
  lines.close();
  txt.close();
print('후추 가리기 완료')


#%% 그림 가리기(coarse salt-10%)
for img in range(n):
  title = random.choice(glob.glob(path));
  image = cv2.imread(title);
  lines = open(str(title[:-4]+ '.txt'), 'r');
  aug = iaa.CoarseSalt(0.1, size_percent=(0.01, 0.1));
  image_aug = aug(image = image);
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'coarse_salt_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'coarse_salt_'+str(img) +'.txt', 'w');
  txt.write(lines.read());
  lines.close();
  txt.close();
print('소금 가리기 완료')


#%% 그림 회전(위아래)
titles = random.sample(glob.glob(path), n);
for img in range(n):
  title = titles[img];
  image = cv2.imread(title);
  texts = [];
  lines = open(str(title[:-4]+ '.txt'), 'r').readlines();
  for line in range(len(lines)):
    lines[line] = lines[line][:-1].split(' ');
  texts.append(lines);
  aug = iaa.Flipud(1.0);
  bbs_aug = [];
  for line in range(len(lines)):
    bbs_aug.append([texts[0][line][0]+' ' ,texts[0][line][1]+' ',str(1.0-float(texts[0][line][2]))+' ',texts[0][line][3]+' ',texts[0][line][4]+'\n']);
  image_aug = aug(image = image);    
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'flipud_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'flipud_'+str(img) +'.txt', 'w');
  for j in range(len(lines)):
    txt.writelines([bbs_aug[j][0], bbs_aug[j][1], bbs_aug[j][2], bbs_aug[j][3], bbs_aug[j][4]]);
  txt.close();
print('상하반전 완료')


#%% 그림 회전(좌우)
titles = random.sample(glob.glob(path), n);
for img in range(n):
  title = titles[img];
  image = cv2.imread(title);
  texts = [];
  lines = open(str(title[:-4]+ '.txt'), 'r').readlines();
  for line in range(len(lines)):
    lines[line] = lines[line][:-1].split(' ');
  texts.append(lines);
  aug = iaa.Fliplr(1.0);
  bbs_aug = [];
  for line in range(len(lines)):
    bbs_aug.append([texts[0][line][0]+' ' ,str(1.0-float(texts[0][line][1]))+' ',texts[0][line][2]+' ',texts[0][line][3]+' ',texts[0][line][4]+'\n']);
  image_aug = aug(image = image);    
  cv2.imwrite(os.path.join('/content/drive/MyDrive/cbe_paper/data_augmt1/', f'fliplr_'+str(img)+'.jpg'), image_aug);
  txt = open('/content/drive/MyDrive/cbe_paper/data_augmt1/'+'fliplr_'+str(img) +'.txt', 'w');
  for j in range(len(lines)):
    txt.writelines([bbs_aug[j][0], bbs_aug[j][1], bbs_aug[j][2], bbs_aug[j][3], bbs_aug[j][4]]);
  txt.close();
print('좌우반전 완료')
```
