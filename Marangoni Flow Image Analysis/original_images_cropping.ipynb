```python
## Original Images Cropping

# Since margins of original images are unmeaningful, I have to crop that parts for the efficiencys of learning

import os

# All files' names in this directory
path = '/content/drive/MyDrive/cbe_paper/rod_particle_marangoni/'
os.chdir(path)
files = os.listdir(path)
print(files)
```

```python
# Image crop

from PIL import Image

new_path = '/content/drive/MyDrive/cbe_paper/rod_particle_marangoni_cropped/'
for image in range(len(files)):
  image_path = path + files[image]
  img = Image.open(image_path)
  area = (165, 0, 1318-165, 741)
  cropped_img = img.crop(area)

  cropped_img_path = new_path + files[image]
  cropped_img.save(cropped_img_path, 'JPEG') 
```
