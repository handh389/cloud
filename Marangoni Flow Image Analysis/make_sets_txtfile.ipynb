###1. Trainset 텍스트 파일 만들기

```python
import os
import glob

# trainset의 이미지들이 담겨있는 폴더의 경로를 아래에 모두 나열한다.
# 단, 마지막엔 *.jpg라 입력하여 jpg 파일(즉, 이미지 파일)만 가져오도록 한다.

train_path1 = '/content/drive/MyDrive/cbe_paper/rod_particle_marangoni_bbx/*.jpg'
train_path2 = '/content/drive/MyDrive/cbe_paper/AgNW_30_001/*.jpg'
train_path3 = '/content/drive/MyDrive/cbe_paper/AgNW_30_002/*.jpg'
train_path4 = '/content/drive/MyDrive/cbe_paper/AgNW_30_005/*.jpg'
train_path5 = '/content/drive/MyDrive/cbe_paper/data_augmt1/*.jpg'
train_path6 = '/content/drive/MyDrive/cbe_paper/data_augmt2/*.jpg'
train_path7 = '/content/drive/MyDrive/cbe_paper/data_augmt3/*.jpg'
train_path8 = '/content/drive/MyDrive/cbe_paper/data_augmt4/*.jpg'

train_path = train_path1, train_path2, train_path3, train_path4, train_path5, train_path6, train_path7, train_path8
```

```python
all_titles = []
for path_num in range(len(train_path)):
  path = train_path[path_num]
  titles = glob.glob(path)
  print(len(titles))
  
  for title_num in range(len(titles)):
    title_name = titles[title_num]
    all_titles.append(title_name)
print(len(all_titles))
```

```python
import random
# 이미지 파일의 순서를 한번 섞어준다.
print(all_titles)
random.shuffle(all_titles)
print(all_titles)
```

```python
## 텍스트 파일 만들기

trainsettext = '/content/drive/MyDrive/cbe_paper/yolotinyv3_medmask_demo/train.txt'
with open(trainsettext, 'w') as f:
  for t in range(len(all_titles)):
    tit = all_titles[t]
    f.write(tit)
    f.write('\n')
  f.close()
```

###2. Testset 텍스트 파일 만들기

```python
import os
import glob

# testset
test_path1 = '/content/drive/MyDrive/cbe_paper/test_rod_particle_marangoni_bbx/*.jpg'
test_path2 = '/content/drive/MyDrive/cbe_paper/test_AgNW_30_001/*.jpg'
test_path3 = '/content/drive/MyDrive/cbe_paper/test_AgNW_30_002/*.jpg'
test_path4 = '/content/drive/MyDrive/cbe_paper/test_AgNW_30_005/*.jpg'


test_path = test_path1, test_path2, test_path3, test_path4
```

```python
all_titles = []
for path_num in range(len(test_path)):
  path = test_path[path_num]
  titles = glob.glob(path)
  print(len(titles))
  
  for title_num in range(len(titles)):
    title_name = titles[title_num]
    all_titles.append(title_name)
print(len(all_titles))
```

```python
import random

print(all_titles)
random.shuffle(all_titles)
print(all_titles)
```

```python
## 텍스트 파일 만들기

testsettext = '/content/drive/MyDrive/cbe_paper/yolotinyv3_medmask_demo/test.txt'
with open(testsettext, 'w') as f:
  for t in range(len(all_titles)):
    tit = all_titles[t]
    f.write(tit)
    f.write('\n')
  f.close()
```
