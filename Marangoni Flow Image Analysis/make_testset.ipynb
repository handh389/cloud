## Trainset과 Testset 분리하기
### - 지구과학교육과 16학번 한동환(handh21@snu.ac.kr)

전체 이미지 중에서 75%는 알고리즘 학습에, 25%는 알고리즘 테스트에 사용하였다. 랜덤으로 이미지를 trainset과 testset으로 나누는 코드이다.

```python
## jpg 파일 로딩

import os
import random

# All jpg files' names in this directory
path = '/content/drive/MyDrive/cbe_paper/rod_particle_marangoni_bbx/'
os.chdir(path)
all_files = os.listdir(path)
print(all_files)

files = []
for file_num in range(len(all_files)):
  file = all_files[file_num]
  if file[-4:] == '.jpg':
    files.append(file)
print(files)


# testset 추출

lenth_files = len(files)
testset_num = int(lenth_files / 4)

testset_files = []
for s in range(testset_num):
  title = random.choice(files)
  testset_files.append(title)
  files.remove(title)

print(testset_files)


# 해당 텍스트 파일
testset_files_text = []
for ff in range(len(testset_files)):
  jpg_file_name = testset_files[ff]
  txt_file_name = jpg_file_name[:-4] + '.txt'
  testset_files_text.append(txt_file_name)

print(testset_files_text)


# 파일 이동 전 확인
print(len(all_files))
print(len(files))
print(len(testset_files))
print(len(testset_files_text))

## 파일 이동시키기
destination_path = '/content/drive/MyDrive/cbe_paper/test_rod_particle_marangoni_bbx/'

for f in range(len(testset_files)):
  f_name = testset_files[f]
  ft_name = testset_files_text[f]
  os.replace(path+f_name, destination_path + f_name)
  os.replace(path+ft_name, destination_path + ft_name)
```
